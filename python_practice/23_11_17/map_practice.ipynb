```python
import pandas as pd
import folium

m = folium.Map(location=[37.547631, 126.942463], zoom_start=15)

data = pd.read_csv("./택시승차대 현황.csv", encoding='cp949')
df=pd.DataFrame(data)

for index, row in df.iterrows():
    folium.Marker([row['위도'], row['경도']], popup='subway', icon=folium.Icon(color='red', icon='fa-solid fa-taxi', prefix='fa-solid')).add_to(m)
m
```
